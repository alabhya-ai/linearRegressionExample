 <b> suppose we have to make a linear regression model to predict house prices in a certain place, given the area of the house.

 tip: linear regression is used to make predictive models when the data is also scattered linearly </b>

```python
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3, 4], [1, 2.5, 4, 5.5])
plt.xlabel("area of house in units of 10m^2 (X)")
plt.ylabel("price of house in units of Rs. 100,000 (Y)")
plt.title("House Prices and Areas")
plt.show()
```

<b> We run Gradient Descent to Find optimal coefficient and intercept in our univariate linear regression model </b>

```python
def GD(a, b, alpha, m, X, Y):
  a = a - ((alpha / m) * sum([(a * Xi + b - Yi) * Xi for Xi, Yi in zip(X, Y)]))
  b = b - ((alpha / m) * sum([(a * Xi + b - Yi) for Xi, Yi in zip(X, Y)]))
  return a, b
```

<b> parameters required while running Gradient Descent </b>

```python
a = 0 # starting values of a and b are commonly kept 0
b = 0
alpha = 0.1 # alpha is usually given a small value
X = [1, 2, 3, 4] # area of house in units of 10m^2
m = len(X) # number of data points
Y = [1, 2.5, 4, 5.5] # price of house in units of Rs. 100,000
```

<b> Applying the Gradient Descent Algorithm </b>

```python
# first iteration
new_a, new_b = GD(a, b, alpha, m, X, Y)
print(new_a, new_b)

# 2nd to (n + 1)th iteration
n = 1000
for i in range(n):
  new_a, new_b = GD(new_a, new_b, alpha, m, X, Y)
  print(new_a, new_b)
```

<b> optimum values of a and b </b>

```python
new_a, new_b = round(new_a, 2), round(new_b, 2) # rounding off to 2 decimal places
print(new_a, new_b)
```

```python
# trained model
f = lambda x: new_a * x + new_b # f(x) = 1.5 * x + (-0.5)
```

```python
# our model predicts house price for a house having area 3 * 10m^2 as approx. Rs. 4 * 100,000
print(f(3))
```

```python
# Our model's predictions on the given data
pred = list(map(f, X))
print(pred)
```

<b> model visualization (THE STRAIGHT RED LINE IS OUR LINEAR REGRESSION MODEL!!)

I highly advise experimenting by changing the number of iterations in gradient descent </b>

```python
plt.scatter(X, Y)
plt.xlabel("area of house in units of 10m^2 (X)")
plt.ylabel("price of house in units of Rs. 100,000 (Y)")
plt.title("House Prices and Areas")
plt.plot(X, pred, color="red")
plt.show()
```

```python
# our model is:
print(f"y = ({new_a})x + ({new_b})")
```
